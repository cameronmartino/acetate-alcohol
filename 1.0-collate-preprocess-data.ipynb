```python
import os
import glob
import pandas as pd
import numpy as np
```

## group samples into one tsv/csv

```python
indir_annot = 'data/funcational-annotations/bin_funct_annotations'
all_gff = {}
for bin_ in glob.glob(os.path.join(indir_annot,'*.gff')):
    bin_name = bin_.split('/')[-1].split('.g')[0]
    annot_ = pd.read_csv(bin_,sep='\t', header=None)
    info_ = pd.read_csv(os.path.join(indir_annot.replace('bin_funct_annotations','prokka_out_tsv'),
                             '%s.tsv'%(bin_name)), sep='\t')
    all_gff[bin_name] = pd.concat([annot_, info_], axis=1)

all_gff_df = pd.concat(all_gff).reset_index().drop('level_1', axis=1)
all_gff_df.columns = all_gff_df.columns.astype(str)
all_gff_df = all_gff_df.rename({'level_0':'bin_id','0':'contig','1':'source',
                               '2':'region','3':'start',
                               '4':'end','5':'score',
                               '6':'strand','7':'phase',
                               '8':'attributes'}, axis=1)
all_gff_df.to_csv('data/funcational-annotations/grouped-annotations.gff', sep='\t')
all_gff_df.head(5)
```

```python
indir_binabund = 'data/shotgun-read-mapping/raw-data'
all_binabund = []
for bin_ in glob.glob(os.path.join(indir_binabund,'*.counts')):
    name_ = bin_.split('/')[-1].split('.')[0]
    all_binabund.append(pd.read_csv(bin_, index_col=0,
                                    sep='\t').rename({'count':name_},
                                                     axis=1))
all_binabunddf = pd.concat(all_binabund, axis=1)
all_binabunddf = all_binabunddf[all_binabunddf.sum(axis=1) > 0]
all_binabunddf.columns = [c_.split('_')[0] for c_ in all_binabunddf.columns]
all_binabunddf.to_csv('data/shotgun-read-mapping/contig-abundance-TPM.tsv', sep='\t')
all_binabunddf.head(5)
    
```

```python
indir_transbund = 'data/transcriptional-mapping/raw-data'
all_transcriptbund = []
for trans_ in glob.glob(os.path.join(indir_transbund,'*.counts')):
    name_ = trans_.split('/')[-1].split('.')[0]
    all_transcriptbund.append(pd.read_csv(trans_, index_col=0,
                                          sep='\t').rename({'count':name_},
                                                           axis=1))
all_transcript_abunddf = pd.concat(all_transcriptbund, axis=1)
all_transcript_abunddf = all_transcript_abunddf[all_transcript_abunddf.sum(axis=1) > 0]
all_transcript_abunddf.columns = [c_.split('_')[0] for c_ in all_transcript_abunddf.columns]
all_transcript_abunddf.to_csv('data/transcriptional-mapping/contig-transcription-TPM.tsv', sep='\t')
all_transcript_abunddf.head(5)
```

## from raw-counts group by each bin and sum the total reads & normalized by bin length length

```python
# bin size in kb
bin_size = {k:[sum([int(x.split('_')[-1]) for x in df.contig])//1000]
            for k, df in all_gff_df.groupby('bin_id')}
bin_size = pd.DataFrame(bin_size,['size_kb']).T
bin_size.to_csv('data/bin-grouped-tables/bin-size.tsv',sep='\t')
# group by bin
bin_abund = []
for bin_, bin_df in all_gff_df.groupby('bin_id'):
    all_binabunddf_bin_counts = all_binabunddf_raw_counts.copy()
    all_binabunddf_bin_counts.index = [x.split('_cov')[0] for x in all_binabunddf_bin_counts.index]
    all_binabunddf_bin_counts = all_binabunddf_bin_counts.groupby(all_binabunddf_bin_counts.index).sum()
    all_binabunddf_bin_counts = all_binabunddf_bin_counts.reindex(set(bin_df.contig)).sum()
    all_binabunddf_bin_counts = pd.DataFrame(all_binabunddf_bin_counts,columns=[bin_]).T
    bin_abund.append(all_binabunddf_bin_counts)    
# raw counts
all_binabunddf_bin_counts_counts = pd.concat(bin_abund, axis=0)
all_binabunddf_bin_counts_counts.to_csv('data/bin-grouped-tables/bin-raw-counts.tsv',sep='\t')
# RPK
all_binabunddf_bin_counts_RPK = (pd.concat(bin_abund, axis=0) /
                                 bin_size.reindex(pd.concat(bin_abund, axis=0).index).values)
all_binabunddf_bin_counts_RPK.to_csv('data/bin-grouped-tables/bin-RPK.tsv',sep='\t')
```

```python
# group by bin
bin_transcript = []
for bin_, bin_df in all_gff_df.groupby('bin_id'):
    all_binabunddf_bin_counts = all_transcript_abunddf_raw_counts.copy()
    all_binabunddf_bin_counts.index = [x.split('_cov')[0] for x in all_binabunddf_bin_counts.index]
    all_binabunddf_bin_counts = all_binabunddf_bin_counts.groupby(all_binabunddf_bin_counts.index).sum()
    all_binabunddf_bin_counts = all_binabunddf_bin_counts.reindex(set(bin_df.contig)).sum()
    all_binabunddf_bin_counts = pd.DataFrame(all_binabunddf_bin_counts,columns=[bin_]).T
    bin_transcript.append(all_binabunddf_bin_counts)    
# raw counts
all_binabunddf_bin_counts_counts = pd.concat(bin_transcript, axis=0)
all_binabunddf_bin_counts_counts.to_csv('data/bin-grouped-tables/bin-transcription-raw-counts.tsv',sep='\t')
# RPK
all_binabunddf_bin_counts_RPK = (pd.concat(bin_transcript, axis=0) /
                                 bin_size.reindex(pd.concat(bin_transcript, axis=0).index).values)
all_binabunddf_bin_counts_RPK.to_csv('data/bin-grouped-tables/bin-transcription-RPK.tsv',sep='\t')
    
```
